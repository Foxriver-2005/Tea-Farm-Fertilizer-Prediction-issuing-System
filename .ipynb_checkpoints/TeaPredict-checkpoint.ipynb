```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('TeaDataset.csv')
dataset
```

```python
dataset.corr()['fertilizer']
dataset.corr()['fertilizer']
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset['bushes'],dataset['fertilizer'])
plt.xlabel('bushes')
plt.ylabel('fertilizer')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset['kilos'],dataset['fertilizer'])
plt.xlabel('kilos')
plt.ylabel('fertilizer')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset['size'],dataset['fertilizer'])
plt.xlabel('size')
plt.ylabel('fertilizer')
plt.show()
```

```python
dataset_np = dataset.to_numpy()
dataset_np.shape
```

```python
X_train, Y_train = dataset_np[:,:3],dataset_np[:,-1]
X_train.shape,Y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train,Y_train)
prediction = model.predict(X_train)
prediction


import pickle
with open('LinearRegression.pkl','wb') as file:
    pickle.dump(model, file)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(prediction,Y_train),mean_squared_error(prediction,Y_train)
```
